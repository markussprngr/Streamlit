## 🎯 Das Grundproblem — bevor wir anfangen

### Die Situation: Eine Party mit 4 Gesprächen gleichzeitig

Stell dir eine Party vor. Es gibt **4 Paare** die gleichzeitig miteinander reden. Jedes Paar hat seinen eigenen Sender (Mund) und Empfänger (Ohr).

Das Problem: Wenn alle gleichzeitig reden, **stören sie sich gegenseitig**. Paar 1 hört nicht nur seinen eigenen Partner, sondern auch Gesprächsfetzen von den anderen 3 Paaren.

Damit Paar 1 seinen Partner trotzdem versteht, muss das **eigene Signal laut genug** sein — im Vergleich zu allem anderen Lärm.

Das ist genau das Problem hier — nur mit Funksendern statt Menschen.

---

### Die Variablen — was bedeutet was?

| Symbol | Name | Bedeutung auf der Party | Wert |
|--------|------|--------------------------|------|
| $p_i$ | Sendeleistung | Wie laut redet Person i? | **gesucht** |
| $G_{ii}$ | Direkter Kanalgewinn | Wie gut hört Empfänger i seinen eigenen Sender? | Diagonale von G |
| $G_{ij}$, $j\neq i$ | Interferenz-Kanalgewinn | Wie stark stört Sender j den Empfänger i? | Rest von G |
| $\sigma_i$ | Rauschen | Wie laut ist die Hintergrundmusik? | gegeben |
| $\gamma_i$ | SINR-Schwelle | Mindest-Verständlichkeit die Paar i braucht | gegeben |

**Wichtig:** $G_{ii}$ (direktes Signal) ist immer **größer** als $G_{ij}$ (Störung). Man hört seinen eigenen Partner besser als die Fremden — das spiegelt die Matrix G wider.

---

### Was ist SINR?

**SINR = Signal-to-Interference-plus-Noise Ratio** = Verhältnis von eigenem Signal zu allem anderen Lärm.

$$\text{SINR}_i = \frac{\overbrace{G_{ii} \cdot p_i}^{\text{eigenes Signal}}}{\underbrace{\sum_{j \neq i} G_{ij} \cdot p_j}_{\text{Störung der anderen}} + \underbrace{\sigma_i}_{\text{Rauschen}}}$$

Damit die Verbindung funktioniert, muss gelten: $\text{SINR}_i \geq \gamma_i$

**Beispiel mit Zahlen:** Angenommen $G_{11} = 1.0$, $p_1 = 0.2$, Störung = 0.05, Rauschen = 0.10:
$$\text{SINR}_1 = \frac{1.0 \times 0.2}{0.05 + 0.10} = \frac{0.2}{0.15} \approx 1.33$$

Wenn $\gamma_1 = 1.2$, dann ist $1.33 \geq 1.2$ ✓ — Verbindung funktioniert.

---

### Das Ziel

Wir wollen die Sendeleistungen $p_1, p_2, p_3, p_4$ so wählen, dass:
- Alle 4 SINR-Bedingungen erfüllt sind (jeder versteht seinen Partner)
- Die **Gesamtleistung minimal** ist (Akku sparen, andere Netze nicht stören)

---
## 🔧 Warum ist das ein LP? — Die Linearisierung

Die SINR-Bedingung hat einen **Bruch** — das ist kein lineares Problem, und LP kann nur lineare Probleme lösen. Wir müssen den Bruch loswerden.

**Schritt für Schritt** (am Beispiel von Nutzer $i=1$):

**Ausgangssituation:**
$$\frac{G_{11} \cdot p_1}{G_{12} p_2 + G_{13} p_3 + G_{14} p_4 + \sigma_1} \geq \gamma_1$$

**Schritt 1:** Nenner einen Namen geben: $D = G_{12} p_2 + G_{13} p_3 + G_{14} p_4 + \sigma_1$. Und $D > 0$ (Leistungen und Rauschen sind positiv).

**Schritt 2:** Beide Seiten mit $D$ multiplizieren (erlaubt weil $D > 0$, Ungleichungsrichtung bleibt):
$$G_{11} \cdot p_1 \geq \gamma_1 \cdot D$$

**Schritt 3:** $D$ ausschreiben:
$$G_{11} p_1 \geq \gamma_1 \cdot (G_{12} p_2 + G_{13} p_3 + G_{14} p_4 + \sigma_1)$$

**Schritt 4:** Rechte Seite ausmultiplizieren, alles mit $p$ auf die linke Seite:
$$G_{11} p_1 - \gamma_1 G_{12} p_2 - \gamma_1 G_{13} p_3 - \gamma_1 G_{14} p_4 \geq \gamma_1 \sigma_1$$

✅ **Jetzt ist es linear in $p$!** Kein Bruch mehr, nur noch Addition und Multiplikation mit Konstanten.

Das Gleiche machen wir für $i=2,3,4$ → 4 Ungleichungen → eine Matrix.

---
## 📐 Die 4 linearen Ungleichungen — mit echten Zahlen

Das ist die Antwort auf **"Derive the equivalent linear inequalities in p ∈ ℝ⁴"**.

### Was bedeutet das auf Deutsch?

"Derive" = herleiten.  
"Equivalent" = gleichwertig (dieselbe Aussage wie die SINR-Bedingung, nur ohne Bruch).  
"Linear inequalities in p" = Ungleichungen wo p nur als $c \cdot p_i$ vorkommt — kein $p^2$, kein Bruch.

### Die Formel allgemein

Für jeden Nutzer $i$ gilt nach der Umformung:
$$G_{ii} \cdot p_i - \gamma_i \sum_{j \neq i} G_{ij} \cdot p_j \geq \gamma_i \cdot \sigma_i$$

### Die 4 Ungleichungen mit den Zahlen aus der Aufgabe

Wir setzen G, gamma und sigma ein:

**Nutzer 1** ($G_{11}=1.0$, $\gamma_1=1.2$, $\sigma_1=0.10$):
$$1.0\, p_1 - 1.2 \cdot 0.1\, p_2 - 1.2 \cdot 0.2\, p_3 - 1.2 \cdot 0.1\, p_4 \geq 1.2 \cdot 0.10$$
$$\Rightarrow \quad 1.0\, p_1 - 0.12\, p_2 - 0.24\, p_3 - 0.12\, p_4 \geq 0.12$$

**Nutzer 2** ($G_{22}=1.2$, $\gamma_2=1.0$, $\sigma_2=0.10$):
$$-1.0 \cdot 0.2\, p_1 + 1.2\, p_2 - 1.0 \cdot 0.1\, p_3 - 1.0 \cdot 0.2\, p_4 \geq 1.0 \cdot 0.10$$
$$\Rightarrow \quad -0.2\, p_1 + 1.2\, p_2 - 0.1\, p_3 - 0.2\, p_4 \geq 0.10$$

**Nutzer 3** ($G_{33}=1.1$, $\gamma_3=1.1$, $\sigma_3=0.15$):
$$-1.1 \cdot 0.1\, p_1 - 1.1 \cdot 0.3\, p_2 + 1.1\, p_3 - 1.1 \cdot 0.2\, p_4 \geq 1.1 \cdot 0.15$$
$$\Rightarrow \quad -0.11\, p_1 - 0.33\, p_2 + 1.1\, p_3 - 0.22\, p_4 \geq 0.165$$

**Nutzer 4** ($G_{44}=1.0$, $\gamma_4=0.9$, $\sigma_4=0.10$):
$$-0.9 \cdot 0.2\, p_1 - 0.9 \cdot 0.1\, p_2 - 0.9 \cdot 0.2\, p_3 + 1.0\, p_4 \geq 0.9 \cdot 0.10$$
$$\Rightarrow \quad -0.18\, p_1 - 0.09\, p_2 - 0.18\, p_3 + 1.0\, p_4 \geq 0.09$$

### Als Matrixform (Kurzschreibweise für alle 4 gleichzeitig)

Die 4 Ungleichungen zusammen schreibt man als:

$$\underbrace{\begin{pmatrix} 1.0 & -0.12 & -0.24 & -0.12 \\ -0.2 & 1.2 & -0.1 & -0.2 \\ -0.11 & -0.33 & 1.1 & -0.22 \\ -0.18 & -0.09 & -0.18 & 1.0 \end{pmatrix}}_{C} \begin{pmatrix} p_1 \\ p_2 \\ p_3 \\ p_4 \end{pmatrix} \geq \underbrace{\begin{pmatrix} 0.12 \\ 0.10 \\ 0.165 \\ 0.09 \end{pmatrix}}_{\gamma \odot \sigma}$$

Das ist genau das `C @ p >= rhs` im Code. **Die Matrix C ist die Herleitung** — jede Zeile ist eine der 4 Ungleichungen.

## 🤖 Was ist cvxpy?

cvxpy ist eine Python-Bibliothek die LP (und andere Optimierungsprobleme) löst. Man beschreibt das Problem — Variablen, Zielfunktion, Constraints — und cvxpy findet die optimale Lösung.

**Analogie:** cvxpy ist wie ein Taschenrechner für Optimierungsprobleme. Du sagst ihm: "Minimiere das, unter diesen Bedingungen" — und er rechnet es aus.

**Grundstruktur immer gleich:**
```python
x = cp.Variable(n)                    # 1. Variablen definieren
prob = cp.Problem(
    cp.Minimize(cp.sum(x)),           # 2. Zielfunktion
    [A @ x <= b, x >= 0]             # 3. Constraints (Liste!)
)
prob.solve()                          # 4. Lösen
print(x.value)                        # 5. Lösung ausgeben
```

````python
# Bibliotheken laden
# numpy: für Matrizen und Vektoren
# cvxpy: für das Optimierungsproblem
# matplotlib: für Plots
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
````

````python
# Die gegebenen Systemparameter aus der Aufgabe

# G ist die Kanalmatrix (4x4):
# G[i][i] = direkter Kanal von Sender i zu Empfänger i (groß, auf der Diagonalen)
# G[i][j] = Störkanal von Sender j zu Empfänger i (klein, außerhalb der Diagonalen)
G = np.array([
    [1.0, 0.1, 0.2, 0.1],   # Empfänger 1: empfängt 1.0 von Sender 1 (gut), 0.1/0.2/0.1 Störung
    [0.2, 1.2, 0.1, 0.2],   # Empfänger 2: empfängt 1.2 von Sender 2 (gut), Rest Störung
    [0.1, 0.3, 1.1, 0.2],   # Empfänger 3
    [0.2, 0.1, 0.2, 1.0]    # Empfänger 4
])

# sigma: Rauschen bei jedem Empfänger (Hintergrundlärm)
sigma = np.array([0.10, 0.10, 0.15, 0.10])

# gamma: SINR-Mindestschwelle für jeden Empfänger
# z.B. gamma[0] = 1.2 bedeutet: Empfänger 1 braucht mindestens SINR = 1.2
gamma = np.array([1.2, 1.0, 1.1, 0.9])

n = 4  # Anzahl Sender-Empfänger-Paare

print("Kanalmatrix G (Diagonale = direkter Kanal, Rest = Störung):")
print(G)
print("\nRauschen sigma:", sigma)
print("SINR-Schwellen gamma:", gamma)
````

## Task 1: Model Construction

### 📋 Aufgabe (Original)
> *For a system with n = 4, use the following data [G, σ, γ]. Derive the equivalent linear inequalities in the power vector **p** ∈ ℝ⁴. Implement the model in Python and verify that your code constructs the constraint matrix correctly.*

### 💡 Erklärung

Wir bauen die Matrix $C$ im Code — das ist die Umsetzung der Herleitung von oben.

Die Einträge von $C$ sind:
- **Diagonale** ($i=j$): $C_{ii} = G_{ii}$ — das eigene Signal, positiv
- **Außerdiagonale** ($i\neq j$): $C_{ij} = -\gamma_i \cdot G_{ij}$ — Störsignal, negativ

**In numpy** trennen wir Diagonale und Außerdiagonale:
- `np.diag(np.diag(G))` → 4×4 Matrix, nur Diagonalelemente, Rest 0
- `G - np.diag(np.diag(G))` → 4×4 Matrix, nur Außerdiagonalelemente, Rest 0
- `np.diag(gamma) @ G_off` → multipliziert jede Zeile $i$ von G_off mit $\gamma_i$

### 🐍 Lösung

````python
# Schritt 1: Diagonale und Außerdiagonale von G trennen
G_diag = np.diag(np.diag(G))
# np.diag(G) holt die Diagonalelemente als Vektor: [1.0, 1.2, 1.1, 1.0]
# np.diag(...) darauf macht daraus wieder eine 4x4 Matrix (Nullen außerhalb)

G_off = G - G_diag
# Außerdiagonale = Gesamtmatrix minus Diagonalmatrix

# Schritt 2: Constraint-Matrix C aufbauen
# np.diag(gamma) = 4x4 Diagonalmatrix mit gamma auf der Diagonalen
# np.diag(gamma) @ G_off = jede Zeile i von G_off wird mit gamma[i] multipliziert
C = G_diag - np.diag(gamma) @ G_off

# Schritt 3: Rechte Seite (gamma_i * sigma_i für jedes i)
rhs = gamma * sigma  # elementweise: [1.2*0.10, 1.0*0.10, 1.1*0.15, 0.9*0.10]

print("Constraint-Matrix C (entspricht der Herleitung oben):")
print(np.round(C, 3))
print("\nRechte Seite gamma * sigma:", np.round(rhs, 4))
````

````python
# Plausibilitätscheck: Diagonale muss positiv sein (eigenes Signal dominiert)
# Außerdiagonale muss negativ sein (Störterme)
diag_C = np.diag(C)
off_diag_vals = C[~np.eye(n, dtype=bool)]
print("Diagonaleinträge von C:", np.round(diag_C, 3))
print("Alle positiv?", np.all(diag_C > 0))
print("Alle Außerdiagonaleinträge negativ?", np.all(off_diag_vals < 0))
````

## Task 2: Minimum Total Power

### 📋 Aufgabe (Original)
> *Solve the optimization problem $\min_{\mathbf{p}\geq 0} \sum_{i=1}^{4} p_i$ subject to all signal to interference plus noise constraints. Your program must: solve the problem with cvxpy, print the optimal power allocation, print the achieved SINR for each user, verify numerically that all constraints are satisfied.*

### 💡 Erklärung

Wir lösen jetzt das eigentliche Optimierungsproblem: **Minimiere die Gesamtleistung aller 4 Sender**, sodass alle SINR-Bedingungen erfüllt sind.

Warum minimale Gesamtleistung?
- In echten Netzen = Akku schonen
- Weniger Sendeleistung = weniger Störung für andere Netze

Das Problem in normaler Sprache:
> *"Finde die kleinstmöglichen Sendeleistungen $p_1, p_2, p_3, p_4$ (alle ≥ 0), sodass jeder Empfänger sein SINR-Minimum erreicht."*

### 🐍 Lösung

````python
# Entscheidungsvariablen: p ist ein Vektor mit 4 Einträgen [p1, p2, p3, p4]
# cvxpy weiß noch nicht welche Werte sie haben — das findet .solve() heraus
p = cp.Variable(n)

# Das Optimierungsproblem definieren
prob = cp.Problem(
    cp.Minimize(cp.sum(p)),    # Zielfunktion: p1 + p2 + p3 + p4 minimieren
    [
        C @ p >= rhs,          # Constraint 1: alle 4 SINR-Ungleichungen (in Matrixform)
        p >= 0                 # Constraint 2: Sendeleistungen können nicht negativ sein
    ]
)

prob.solve()

p_opt = p.value  # die optimalen Sendeleistungen
print(f"Status: {prob.status}")  # 'optimal' = Lösung gefunden
print(f"Minimale Gesamtleistung: {prob.value:.4f}")
print(f"Optimale Leistungen [p1, p2, p3, p4]: {np.round(p_opt, 4)}")
````

````python
# Verifikation: Hat jeder Empfänger wirklich sein SINR-Minimum erreicht?
# Wir rechnen SINR für jeden Nutzer manuell nach — aus der Originalformel (mit Bruch)
print(f"{'Nutzer':>8} {'SINR erreicht':>15} {'SINR benötigt':>15} {'OK?':>6}")
print("-" * 48)

for i in range(n):
    eigenes_signal = G[i, i] * p_opt[i]
    stoerung = sum(G[i, j] * p_opt[j] for j in range(n) if j != i)
    sinr = eigenes_signal / (stoerung + sigma[i])
    ok = sinr >= gamma[i] - 1e-5
    print(f"{i+1:>8} {sinr:>15.4f} {gamma[i]:>15.1f} {'✓' if ok else '✗':>6}")
````

````python
fig, ax = plt.subplots(figsize=(7, 4))
balken = ax.bar([f'Sender {i+1}' for i in range(n)], p_opt,
                color='steelblue', edgecolor='black')
for balken_i, wert in zip(balken, p_opt):
    ax.text(balken_i.get_x() + balken_i.get_width() / 2,
            balken_i.get_height() + 0.002,
            f'{wert:.3f}', ha='center', va='bottom', fontsize=10)
ax.set_xlabel('Sender')
ax.set_ylabel('Sendeleistung $p_i$')
ax.set_title(f'Optimale Leistungsverteilung (Gesamt = {prob.value:.3f})')
ax.grid(axis='y', alpha=0.4)
plt.tight_layout()
plt.show()
````

## Task 3: Power Limits and Feasibility

### 📋 Aufgabe (Original)
> *Add upper bounds $0 \leq p_i \leq p_{\max}$ for all $i$. Investigate feasibility as a function of $p_{\max}$: test a range of values, determine the smallest value for which the problem becomes feasible, visualize feasibility over the tested range, report one feasible and one infeasible case.*

### 💡 Erklärung

Bisher konnten die Sender so laut senden wie nötig — unbegrenzt. In der Realität gibt es aber ein **Hardware-Limit**: Ein Handy kann nicht beliebig viel Leistung senden.

Wir fügen die Bedingung $p_i \leq p_{\max}$ hinzu und fragen: Wenn das Limit sehr klein ist — ist das Problem noch lösbar? Ab welchem $p_{\max}$ wird es lösbar?

**Intuition:** Wenn alle Sender maximal 0.01 senden dürfen, aber für die SINR-Bedingung 0.2 gebraucht werden → **infeasible** (unlösbar). Sobald $p_{\max}$ groß genug ist → **feasible** (lösbar).

### 🐍 Lösung

````python
# Bereich gewählt weil p_opt.max() ≈ 0.3 → alles über 0.3 sollte machbar sein
p_max_werte = np.linspace(0.01, 1.0, 100)
machbar = []

for p_max in p_max_werte:
    p_var = cp.Variable(n)
    prob_lim = cp.Problem(
        cp.Minimize(cp.sum(p_var)),
        [C @ p_var >= rhs, p_var >= 0, p_var <= p_max]  # neu: oberes Limit
    )
    prob_lim.solve(solver=cp.CLARABEL)
    machbar.append(prob_lim.status == 'optimal')

machbar = np.array(machbar)
p_min_machbar = p_max_werte[np.argmax(machbar)]  # erster True-Wert

print(f"Kleinster machbarer p_max:        {p_min_machbar:.3f}")
print(f"Größte Einzelleistung ohne Limit: {p_opt.max():.3f}")
print(f"\nBeispiel nicht lösbar: p_max = {p_max_werte[5]:.3f}")
print(f"Beispiel lösbar:       p_max = {p_max_werte[-1]:.3f}")
````

````python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(p_max_werte, machbar.astype(int), 'b-', linewidth=2)
ax.axvline(p_min_machbar, color='red', linestyle='--',
           label=f'Kleinster machbarer p_max ≈ {p_min_machbar:.3f}')
ax.text(p_min_machbar / 2, 0.5, 'nicht lösbar', ha='center', fontsize=10, color='red')
ax.text((p_min_machbar + 1.0) / 2, 0.5, 'lösbar', ha='center', fontsize=10, color='blue')
ax.set_xlabel('p_max (maximale Sendeleistung pro Sender)')
ax.set_ylabel('Lösbar (1 = ja, 0 = nein)')
ax.set_title('Machbarkeit als Funktion des Leistungslimits p_max')
ax.set_yticks([0, 1])
ax.legend()
ax.grid(True, alpha=0.4)
plt.tight_layout()
plt.show()
````

## Task 4: Uniform QoS Scaling

### 📋 Aufgabe (Original)
> *Replace the threshold vector γ by αγ, where α > 0 is a scalar quality of service factor. Write a program that determines the largest value of α such that the constrained power control problem remains feasible under a fixed power limit $p_i \leq 2$. Present: the estimated maximal α, the corresponding power allocation, a plot showing feasibility as a function of α.*

### 💡 Erklärung

**QoS = Quality of Service = Dienstgüte.** Der Faktor $\alpha$ skaliert alle SINR-Anforderungen gleichzeitig:
- $\alpha = 1$: originale Anforderungen
- $\alpha = 2$: alle Anforderungen doppelt so streng → mehr Leistung nötig
- $\alpha = 0.5$: alle Anforderungen halb so streng → weniger Leistung nötig

**Frage:** Wenn jeder Sender maximal $p_i \leq 2$ senden darf — wie hoch können wir $\alpha$ schrauben, bevor das Problem unlösbar wird?

### 🐍 Lösung

````python
alpha_werte = np.linspace(0.1, 5.0, 200)
p_max_qos = 2.0
machbar_alpha = []
gesamtleistung_alpha = []

for alpha in alpha_werte:
    gamma_a = alpha * gamma          # SINR-Anforderungen skalieren
    C_a   = G_diag - np.diag(gamma_a) @ G_off
    rhs_a = gamma_a * sigma

    p_var = cp.Variable(n)
    prob_a = cp.Problem(
        cp.Minimize(cp.sum(p_var)),
        [C_a @ p_var >= rhs_a, p_var >= 0, p_var <= p_max_qos]
    )
    prob_a.solve(solver=cp.CLARABEL)
    ok = prob_a.status == 'optimal'
    machbar_alpha.append(ok)
    gesamtleistung_alpha.append(prob_a.value if ok else np.nan)

machbar_alpha = np.array(machbar_alpha)
# np.where(machbar_alpha)[0] = alle Indizes wo True → [-1] = letzter
alpha_max = alpha_werte[np.where(machbar_alpha)[0][-1]]
print(f"Maximales α bei p_i ≤ {p_max_qos}: α_max ≈ {alpha_max:.3f}")
print(f"Das bedeutet: SINR-Anforderungen können um Faktor {alpha_max:.2f} erhöht werden.")
````

````python
# Optimale Leistungen beim maximalen α
gamma_best = alpha_max * gamma
C_best = G_diag - np.diag(gamma_best) @ G_off
p_best_var = cp.Variable(n)
cp.Problem(
    cp.Minimize(cp.sum(p_best_var)),
    [C_best @ p_best_var >= gamma_best * sigma, p_best_var >= 0, p_best_var <= p_max_qos]
).solve()
print(f"Leistungen bei α_max = {alpha_max:.2f}: {np.round(p_best_var.value, 4)}")
````

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(alpha_werte, machbar_alpha.astype(int), 'b-', linewidth=2)
ax1.axvline(alpha_max, color='red', linestyle='--', label=f'α_max ≈ {alpha_max:.2f}')
ax1.set_xlabel('α (QoS-Skalierungsfaktor)')
ax1.set_ylabel('Lösbar (1 = ja, 0 = nein)')
ax1.set_title('Machbarkeit vs. α')
ax1.set_yticks([0, 1])
ax1.legend()
ax1.grid(True, alpha=0.4)

ax2.plot(alpha_werte, gesamtleistung_alpha, 'g-', linewidth=2)
ax2.axvline(alpha_max, color='red', linestyle='--', label=f'α_max ≈ {alpha_max:.2f}')
ax2.set_xlabel('α')
ax2.set_ylabel('Minimale Gesamtleistung')
ax2.set_title('Je größer α, desto mehr Leistung nötig')
ax2.legend()
ax2.grid(True, alpha=0.4)

plt.tight_layout()
plt.show()
````

## Task 5: Random Instance Generator

### 📋 Aufgabe (Original)
> *Write a function that generates random wireless network instances with: positive direct gains $G_{ii}$, smaller interference gains $G_{ij}$ for $i \neq j$, positive noise levels, random threshold values. Use this generator to create and solve at least five random instances. For each instance, report: whether the problem is feasible, the minimum total power if feasible, the number of active upper bound constraints if power limits are included.*

### 💡 Erklärung

Wir testen unser Modell mit **zufälligen Netzwerken** — um zu sehen ob es generell funktioniert, nicht nur für das eine Beispiel aus der Aufgabe.

Anforderungen:
- Direktkanäle $G_{ii}$ groß (0.8–1.5) — realistisch: eigener Kanal ist stark
- Interferenzkanäle $G_{ij}$ kleiner (0.05–0.3) — realistisch: Störung ist schwächer

**Aktive Limits** = wie viele Sender senden genau auf dem Maximum $p_{\max}$? 0 = viel Spielraum, 4 = alle senden auf Anschlag.

**Warum seed?** Gleicher seed → immer gleiche Zufallszahlen → Ergebnis reproduzierbar.

### 🐍 Lösung

````python
def generate_instance(n, seed=None):
    rng = np.random.default_rng(seed)
    G_rand = rng.uniform(0.05, 0.3, (n, n))             # schwache Interferenzkanäle
    np.fill_diagonal(G_rand, rng.uniform(0.8, 1.5, n))  # starke Direktkanäle überschreiben
    return G_rand, rng.uniform(0.05, 0.2, n), rng.uniform(0.5, 1.5, n)


def loesung_berechnen(G_r, sigma_r, gamma_r, p_max=None):
    n_r = len(gamma_r)
    G_d = np.diag(np.diag(G_r))
    C_r   = G_d - np.diag(gamma_r) @ (G_r - G_d)
    rhs_r = gamma_r * sigma_r
    p_v = cp.Variable(n_r)
    cons = [C_r @ p_v >= rhs_r, p_v >= 0]
    if p_max is not None:
        cons.append(p_v <= p_max)
    prob = cp.Problem(cp.Minimize(cp.sum(p_v)), cons)
    prob.solve(solver=cp.CLARABEL)
    aktiv = int(np.sum(p_v.value >= p_max - 1e-4)) if prob.status == 'optimal' and p_max else None
    return prob.status, prob.value, aktiv


p_max_test = 2.0
print(f"{'Instanz':>9} {'Status':>15} {'Gesamtleistung':>16} {'Aktive Limits':>15}")
print("-" * 60)
for seed in range(5):
    G_r, sigma_r, gamma_r = generate_instance(n=4, seed=seed)
    status, gesamtleistung, aktiv = loesung_berechnen(G_r, sigma_r, gamma_r, p_max=p_max_test)
    if status == 'optimal':
        print(f"{seed+1:>9} {'lösbar':>15} {gesamtleistung:>16.4f} {aktiv:>15}")
    else:
        print(f"{seed+1:>9} {'nicht lösbar':>15} {'—':>16} {'—':>15}")
````
